```python
plt.figure(figsize=(10, 6))
plt.hist(fuel_econ['co2'], bins=20, color='skyblue', edgecolor='black')
plt.title('Histogram of CO2 Emissions')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(fuel_econ['displ'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Displacement (L)')
plt.ylabel('Frequency')
plt.title('Histogram of Displacements')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(fuel_econ['city'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('City Fuel Efficiency (MPG)')
plt.ylabel('Frequency')
plt.title('Histogram of City Fuel Efficiency')
plt.grid(True)
plt.show()
```

```python
numeric_columns = fuel_econ.select_dtypes(include=['float64', 'int64'])

# Calculating the correlation matrix
corr_matrix = numeric_columns.corr()

# Plotting the heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap of Fuel Economy Data')
plt.show()
```
